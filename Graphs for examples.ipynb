# This is just a notebook containing the graphs used in the examples/ Quizes

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Rubber noise quiz

```python
t = np.linspace(0.01, 0.5, 10)
t_prime = t/0.3*10
sound = 100*((-t**3 + 1.29*t**2 - 0.55*(t) - 0.2 - 0.005*np.random.randn(t.size))+0.26)
ax = sns.regplot(t_prime, sound, scatter_kws={"s": 20}, order=3, ci=90)
ax.set_ylabel('Noise level (Db)')
ax.set_xlabel('Rubber thickness (mm)')
ax.set_xlim([0,None])
ax.grid()
fig = plt.gcf()
fig.savefig('Rubber friction.png', dpi = 300)
```

```python
t = np.linspace(0.01, 0.5, 10)
t_prime = t/0.3*10
ax = sns.regplot(t_prime, sound, scatter_kws={"s": 20}, order=3, ci=90, label='Original')

sound1 = 90*((-1.4*t**3 + 1.27*t**2 - 0.6*(t) - 0.2 - 0.005*np.random.randn(t.size))+.28)
ax = sns.regplot(t_prime, sound1, scatter_kws={"s": 20}, order=3, ci=90, label='Option 1')

sound2 = 100*(- 0.005*np.random.randn(t.size))-2
ax = sns.regplot(t_prime, sound2, scatter_kws={"s": 20}, order=3, ci=90, label='Option 2' )
ax.set_xlim([0,17])
ax.set_ylabel('Noise level (Db)')
ax.set_xlabel('Rubber thickness (mm)')
ax.grid()
ax.legend()
fig = plt.gcf()
fig.savefig('Rubber friction2.png', dpi = 300)
```
