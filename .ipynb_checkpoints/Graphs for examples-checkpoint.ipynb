# This is just a notebook containing the graphs used in the examples/ Quizes

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Rubber friction quiz

```python
speed = np.linspace(0.01, 0.5, 10)
mu = speed**3 - 1.29*speed**2 + 0.5*(speed) + 0.2 + 0.004*np.random.randn(speed.size)
ax = sns.regplot(speed, mu, scatter_kws={"s": 20}, order=3, ci=90)
ax.set_ylabel('Friction coefficient')
ax.set_xlabel('Sliding speed m/s')
fig = plt.gcf()
fig.savefig('Rubber friction.png', dpi = 300)
```
