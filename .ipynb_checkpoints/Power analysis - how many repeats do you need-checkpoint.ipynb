# Power analysis
---
For any design of experiment there is a statistiacl test that you will use to test your results. These tests have assumptions built in concerning the form of the results. These are covered in the statistical tests notebook.

Once you have decided on the statistical test that you will use you can work out the power of the experiment. This is how likely it is to find a positive result (eg. your coating importves wear life, or degrading oil befroe the test increases wear). To work this out we need to define some terms. 

Firstly we say we have found a positive result if the p value of the statistical test is below a certain threshold. (meaning thst the actual out come would be unlikely by chance).

We also need to define what we mean by a positive result. If we are interested in big changes, these are easier to find than small changes (eg a 1% change in wear life would be harder to prove than a 1000% percent change in wear life beacuse the tests we do would all show bigger changes). It turns out that the absolute difference is not important, rather the ratio of the difference we are trying to measure to the standard deviation of the population. Practically this means that it is easier to find differences in very repetable system than system with a lot of variation. These two factors are summarised as the effect size.

Lastly, to calculate the power of the experiemnt we need to know how many repeats will be used. Large experiments with lots of repeats are more likely to find positive results if they are there. 

Practically we often want to calculate the number of repeats which we would need to do in order to achive a set power. This is alos possible and in many cases trivial. 

---
Below, all figures relating to power in the presentation are generated and some examples are worked through.

First we will just import some useful python packages, the most important one here is statsmodels.stats.power, this will do all the power analysis, the rest are just for plotting the results. 

```python
import statsmodels.stats.power
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# The t-test
The t-test is often used to compare the means of two groups (eg a group with a coating and one without a coating, or one at low speed and one at high speed). This is the most simple statistical test you can do. It's asumptions are simple and often met in practice.

---

To calculate the power we can use the function tt_ind_solve_power. This funciton requires three of the four parameters described above (number of repeats, effect size, threshold p value, power) and it will return the remaining parameter.

---
In the example below we will use this function to find the number of repeats which we need to do to have an 80% chance of finding a result if one is present in our data. In this example we will use a threshold p value of 0.05, this is the most common value for this parameter, although it is arbitarily chosen.

The most difficult part of this process is estimating the effect size. Sometimes industry partners will have an idea of the size of change they are interested in. For example, they may only be interested in applying an expensive coating if it doubles the wear life of the component, however, often you will have to estimate the sort of values you are interested in.

The other term in the effect size, the standard deviation of the population is slightly more simple. This can be estimated from relevent wear test standards or your own experience of the system. 

```python
# Finding the effect size
mean_control = 1000 # eg. cycles to failure 
mean_treatment = 2000 # cycles to failure, the minimum change we are interested in finding
standard_deviation_control_group = 300 # cycles to failure
effect_size = (mean_treatment-mean_control)/standard_deviation_control_group
# Threshold p value
p_value = 0.05
# power
power = 0.8
repeats = statsmodels.stats.power.tt_ind_solve_power(effect_size=effect_size, 
                                                     nobs1=None, 
                                                     alpha=p_value, 
                                                     power=power)
print(f'We need to run {int(np.round(repeats))} repeat tests in each group.')
```

So in this example we would need to run 3 repeats to have an 80% chance of finding a positive result if one is there. If it turns out that the change caused by our tratment is larger than we have assumed we will have a better chance of finding it. 

In tribology we are often only interested in large changes so we often only need a few repeats. However you should be aware that for small number of repeats the assumptions of the t-test are not garanteed.

You can play around with the numbers in the code above to see how this influnces the result. 

*note: that the t-test will not work with fewer than 3 repeats so if the answer would be less than 3 the function will not converge and will just return 10 to be safe*

---

In the section below we will plot the ficure from the presetation, this just shows how the power changes if we increase the number of repeats or change our threshold for a positive result (the threshold p value).

Here we have assumed an effect size of 2 which is in line with the median result we found as part of our literature review into experimental design in tribology.

```python
es = 2 # effect size
p_005 = []
p_001 = []
p_0001 = []
for i in range(3,11):
    p_005.append(statsmodels.stats.power.tt_ind_solve_power(effect_size=es, nobs1=i, alpha=0.05, 
                                               power=None, ratio=1.0, alternative='two-sided'))
    p_001.append(statsmodels.stats.power.tt_ind_solve_power(effect_size=es, nobs1=i, alpha=0.01, 
                                               power=None, ratio=1.0, alternative='two-sided'))
    p_0001.append(statsmodels.stats.power.tt_ind_solve_power(effect_size=es, nobs1=i, alpha=0.001, 
                                               power=None, ratio=1.0, alternative='two-sided'))
data = pd.DataFrame({'p = 0.05':p_005, 'p = 0.01':p_001,'p = 0.001':p_0001})
data.index = range(3,11)
ax = data.plot()
ax.set_xlim(3,10)
ax.set_xlabel('Number of repeats (n)')
ax.set_ylim(0,1)
ax.set_ylabel(r'Power ($\beta$)')
```

# The ANOVA
The next most simple design of experiment is to compare several mean. This is tested with an ANOVA. Here we can find the power by using the method solve_power in the FTestAnovaPower class. This is also in the statsmodels package. 

For an ANOVA the number of groups we will test is also important, this is beacause the ANOVA automatically corrects for multiple comparisons. 

---

In the example below we will find the power of the same study that we used for the t-test example but noe we will include more groups in the analysis.

Again the effect size of interest has to be estimted from the size of the change that we are interested in and the expected variation in our results.

Here we will test 4 different groups (eg 3 treatments and one control), these could be different test conditions such as addatives in oil, coatings or process parameters.

In the code below we will plot out the power of the study for different numbers of repeats. ** Note: the ANOVA will not work if the number of repeats in each group is less than or equal to the number of groups. ** Practically this causes a problem as we often want to test a variety of different procuts. In these cases, screening tests can be useful to limit the size of the experiment

```python
# Finding the effect size
mean_control = 1000 # cycles to failure 
mean_treatment = 2000 # cycles to failure
standard_deviation_control_group = 300 # cycles to failure
effect_size = (mean_treatment-mean_control)/standard_deviation_control_group
# Threshold p value
p_value = 0.05
# power
power = 0.8
# number of groups
groups = 4

p = statsmodels.stats.power.FTestAnovaPower()
all_powers = []
for i in range(5, 12):
    power = p.solve_power(effect_size=effect_size, nobs=i, alpha=p_value, 
                            power=None, k_groups = groups,)
    all_powers.append(power)
all_powers = np.array(all_powers)
all_powers[np.isnan(all_powers)] = 0
plt.plot(range(5,12), all_powers)
ax = plt.gca()
ax.set_xlabel('Number of repeats')
ax.set_ylabel('Power')
ax.grid()
```

So in this example we would have to do 7 repeats to get the power we want. This means doing a much larger experiment than the t-test example even though we are interested in the same sort of chnges in the same system. In these cases it can be good to limit the scope of the experiments by using screening tests or testing fewer variants.

---

Below we will create the graph shown in the presentation, showing how power changes with the number of groups we want to test and our threshold for a positive result.

```python
results = []
p = statsmodels.stats.power.FTestAnovaPower()
for i in range(3,16):
    for alpha in [0.05, 0.01, 0.001]:
        for k_groups in [2,3,5]:
            
            power = p.solve_power(effect_size=es, nobs=i, alpha=alpha, 
                                                power=None, k_groups = k_groups)
            result = [i, f'p = {alpha}', k_groups, power]
            results.append(result)

results = np.array(results)
power = np.array(results[:,-1], dtype=float)
power[np.isnan(power)] = 0

results = {'n':np.array(results[:,0],dtype=int), 'p':results[:,1], 'k_groups':results[:,2], 'power':power}
data = pd.DataFrame(results, columns=['n', 'p', 'k_groups', 'power'])
ax = sns.lineplot(x='n', y='power', hue='p', style='k_groups', data=data)
ax.set_xlim(3,15)
ax.set_xlabel('Number of repeats (n)')
ax.set_ylim(0,1)
ax.set_ylabel(r'Power ($\beta$)')

plt.gcf().savefig('anova_power.png', dpi = 600)
```

# Linear regression
Power analysis for linear regression is more difficult. Here we need to do monte carlo simulations of data to get an idea of the power of the experiment. This is quite a simple process, all we have to do is generate data with the minimum trend we are interested in, with the varience that we expect, over the range that we will test in. We then fit the model we wil use to the data and see if we have found the trend. If we repeat this enough times we will get an idea of the power of the study.

Below is an example of this process. Again the size of trend that you are interested in and the varience of the residuals (the difference between your trend and the actual results after fitting), need to be estimated. 

In this example we will try to test for a linear trend between force and temperature, we are only interested in a trend of at least 0.05 degrees per newton, and we will test 4 different levels of force between 100 and 500N.

We think that the standard deviation in our temperature readings is roughly 1.5 degrees (if we tested the same system lots of times at the same force this would be the variation).

```python
import statsmodels.api as sm

force = np.linspace(100,500,4)
X = sm.add_constant(force)
num_samples_monte_carlo = 2000
num_found = 0
threshold_pvalue = 0.01

for i in range(num_samples_monte_carlo):
    temperature = 0.05 * force + 1.5 * np.random.randn(force.size)
    model = sm.OLS(temperature, X)
    results = model.fit()
    if results.pvalues[1]<threshold_pvalue:
        num_found += 1
        
print(f'Our estimated power is {num_found/num_samples_monte_carlo}')
```
