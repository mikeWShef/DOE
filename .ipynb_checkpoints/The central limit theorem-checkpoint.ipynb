# The central limit theorem
The central limit theormem states that if we add random values together, no matter how the values are distributed, the distribution of the sum will approach a normal distribution as the number of items in the sum approaches infinity. 

This isa bit confusing so we will just show a simple example. We will take uniformly distributed random numbers and add them together in groups of n. If we do this enough times we can get an idea of the distribution of the numbers, so we will do it a lot.

```python
import numpy as np
import matplotlib.pyplot as plt
n_samples = 50000

totals = np.zeros(n_samples)

fig, ax = plt.subplots(1,4, figsize=(10,4))

plt_num = 0

for i in range(10):
    rvs = np.random.rand(n_samples)
    totals += rvs
    
    if i in [0,1,9]:
        ax[plt_num].hist(totals, bins = 30)
        plt_num += 1
        
a=ax[3].hist(np.random.randn(n_samples), bins = 30)
```

 In the first plot we are showing the uniform distribution, the second shows the distribution of the sum of two uniformly distributed random variables, the third shows the distribution of 10 uniformly distributed random variables. The last figure shows a normal distribution.

As shown the sum of even just 10 uniformly distributed random variables is very similar to the normal distribution. This is in fact true for any continuous random variable. 

This theorem is important for the assumptions of the t-test.S
