```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
x = np.arange(20)+1
```

```python
y1 = 1- 0.95**x
y2 = 1- 0.99**x
y3 = 1- 0.999**x
```

```python
ax = sns.lineplot(x,y1, label='p=0.05')
ax = sns.lineplot(x,y2, label='p=0.01')
ax = sns.lineplot(x,y3, label='p=0.001')
ax.set_xlim(1,20)
ax.set_xlabel('Number of comparisons')
ax.set_ylim(0,0.7)
ax.set_ylabel('False positive rate')
ax.set_xticks([1,2,3,5,10,15,20])
ax.grid()
plt.gcf().savefig('multiple_comparisons.png', dpi=600)
```
